```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = "http://quotes.toscrape.com/page/1/"
```

```python
response = requests.get(url)
```

```python
response.content
```

```python
soup = BeautifulSoup(response.content,'html.parser')
```

```python
soup
```

```python
print(soup.prettify())
```

```python
quote_blocks = soup.find_all('div', class_='quote')
```

```python
names = []
quotes = []
tags = []
```

```python
for block in quote_blocks:
    name = block.find('small', class_='author').get_text()
    names.append(name)
print(names)
```

```python
for block in quote_blocks:
  quote = block.find('span', class_='text').get_text()
  quotes.append(quote)
print(quotes)
```

```python
for block in quote_blocks:
  tag_elements = block.find_all('a', class_='tag')
  tag_list = [tag.get_text() for tag in tag_elements]

  tags.append(tag_list)
print(tags)
```

```python
for block in quote_blocks:
    name = block.find('small', class_='author').get_text()
    quote = block.find('span', class_='text').get_text()
    tag_elements = block.find_all('a', class_='tag')
    tag_list = [tag.get_text() for tag in tag_elements]

    names.append(name)
    quotes.append(quote)
    tags.append(tag_list)


print("Names:")
print(names)
print("\nQuotes:")
print(quotes)
print("\nTags:")
print(tags)
```
